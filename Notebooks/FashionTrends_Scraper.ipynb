```python
def extract_base_url(url):                                                          # function to extract base url
    parsed_url = urlparse(url)
    base_url = parsed_url.scheme + "://" + parsed_url.netloc
    return base_url
```

```python
def check_base_url(url, str):                                                       # function to check whether the prefix is in the base_url
    return str in url
```

```python
# getting Vogue India Fashion URL
try:
    from googlesearch import search
except ImportError:
    print("No module named 'google' found")


fashion_url_list = []
query = "vogue india fashion"
count = 0
for j in search(query, tld="com", lang="en", num=5, pause=2):
    base = extract_base_url(j)
    if(check_base_url(base, "vogue")):
      fashion_url_list.append(j)
      count += 1
    if count == 2:                                                                  # precautionary counter
      break
```

```python
def remove_numbers(sentence):                                                       # function to remove numbers
    return re.sub(r'\d+', '', sentence)
```

```python
def remove_stopwords(content):                                                      # function to remove stopwords
    if content == None:
      return

    words = content.lower().split()
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word not in stop_words]
    cleaned_text = ' '.join(filtered_words)
    return cleaned_text
```

```python
def removal(s):                                                                      # function for preprocessing

    # Change any white space to one space
    stripped = re.sub('\s+', ' ', s)

    # Remove start and end white spaces
    stripped = stripped.strip()
    if stripped != None:
            return stripped.lower()
```

```python
# extracting text content from Vogue India URL
contents_list = []

for url_ in fashion_url_list:
  reqs = requests.get(url_)
  soup = BeautifulSoup(reqs.text, 'html.parser')

  content_list = []
  for content in soup.find_all('h2'):
    content_list.append(remove_numbers(remove_stopwords(removal(content.get_text().lower().strip()))))         # to remove numbers, stopwords and punctuation
  unique_elements_set = set(content_list)                                             # to remove the duplicate elements
  content_list = list(unique_elements_set)
  content_string = str.join(' ', content_list)
  # print(content_string)
  contents_list.append(content_string)

for l in contents_list:
  print(l)
```

```python
from nltk import bigrams, word_tokenize
```

```python
for content in contents_list:
  # Tokenize content into words
  words = word_tokenize(content)

  # Generate bigrams from the list of words
  bigrams_list = list(bigrams(words))

  for bigram in bigrams_list:
      print(bigram)
```

```python
# extracting image content from URLs
for url_ in fashion_url_list:
  reqs = requests.get(url_)
  soup = BeautifulSoup(reqs.text, 'html.parser')
  image_urls = []
  for img_tag in soup.find_all('img'):                                              # targeting img tags
    img_url = img_tag.get('src')
    if img_url:
        if img_url.startswith('http'):
            image_urls.append(img_url)
        else:
            image_urls.append(url_ + img_url)
```

```python
def imgs_from_website(image_urls, save_folder, siz):
    # Create the save folder if it doesn't exist
    os.makedirs(save_folder, exist_ok=True)

    # Initialize a count to keep track of downloaded images
    count = 0

    # Loop through the image URLs
    for i, image_url in enumerate(image_urls):
        # Get the image using the URL
        response = requests.get(image_url)

        # Check if the response is successful (status code 200)
        if response.status_code == 200:
            # Get the image extension from the URL
            image_extension = image_url.split('.')[-1]

            # Set the extension to 'jpg' if it's not 'jpg' or 'jpeg'
            if image_extension != "jpg" or image_extension != "jpeg":
                image_extension = "jpg"

            # Create a filename for the image
            filename = f"image_{i}.{image_extension}"
            image_path = os.path.join(save_folder, filename)

            # Save the image content to the specified path
            with open(image_path, 'wb') as f:
                f.write(response.content)

            # Print a message indicating the image has been saved
            print(f"Image {i + 1}/{len(image_urls)} saved as {filename}")
            count += 1
        else:
            # Print a message if the image download fails
            print(f"Failed to download image {i + 1}: {image_url}")

        # Break the loop if the desired number of images have been downloaded
        if count == siz:
            break
```

```python
#to save session cookie from Firefox to login for instascraper
from argparse import ArgumentParser
from glob import glob
from os.path import expanduser
from platform import system
from sqlite3 import OperationalError, connect

try:
    from instaloader import ConnectionException, Instaloader
except ModuleNotFoundError:
    raise SystemExit("Instaloader not found.\n  pip install [--user] instaloader")


def get_cookiefile():
    default_cookiefile = {
        "Linux": "/home/zeuidon/snap/firefox/common/.mozilla/firefox/2ngefm2z.default/cookies.sqlite",
        "Darwin": "~/Library/Application Support/Firefox/Profiles/*/cookies.sqlite",
    }.get(system(), "~/.mozilla/firefox/*/cookies.sqlite")
    cookiefiles = glob(expanduser(default_cookiefile))
    if not cookiefiles:
        raise SystemExit("No Firefox cookies.sqlite file found. Use -c COOKIEFILE.")
    return cookiefiles[0]


def import_session(cookiefile, sessionfile):
    print("Using cookies from {}.".format(cookiefile))
    conn = connect(f"file:{cookiefile}?immutable=1", uri=True)
    try:
        cookie_data = conn.execute(
            "SELECT name, value FROM moz_cookies WHERE baseDomain='instagram.com'"
        )
    except OperationalError:
        cookie_data = conn.execute(
            "SELECT name, value FROM moz_cookies WHERE host LIKE '%instagram.com'"
        )
    instaloader = Instaloader(max_connection_attempts=1)
    instaloader.context._session.cookies.update(cookie_data)
    username = instaloader.test_login()
    if not username:
        raise SystemExit("Not logged in. Are you logged in successfully in Firefox?")
    print("Imported session cookie for {}.".format(username))
    instaloader.context.username = username
    instaloader.save_session_to_file(sessionfile)


if __name__ == "__main__":
    p = ArgumentParser()
    p.add_argument("-c", "--cookiefile")
    p.add_argument("-f", "--sessionfile")
    args = p.parse_args()
    try:
        import_session(args.cookiefile or get_cookiefile(), args.sessionfile)
    except (ConnectionException, OperationalError) as e:
        raise SystemExit("Cookie import failed: {}".format(e))
```

```python

"""**Scraping Instagram**"""

import instaloader
import datetime
from PIL import Image
import os

# server-side user
USER = 'itesting5678'
loader = instaloader.Instaloader()
# loader.load_session_from_file(USER)

def scrape_images(username):

    try:
        # Load the profile using the given username
        profile = instaloader.Profile.from_username(loader.context, username)

        # Create a folder
        parent_directory = "/home/zeuidon/Documents/flip"
        enhanced_folder = "imageSet_Insta"
        folder_path = os.path.join(parent_directory, enhanced_folder)
        os.makedirs(folder_path, exist_ok=True)

        # The Threshold Date
        current_date = datetime.date.today()
        days_to_subtract = datetime.timedelta(days=7)
        date_threshold = current_date - days_to_subtract

        # Track the processed dates
        processed_dates = set()
        count = 0
        # Iterate over the profile's posts
        for post in profile.get_posts():
            post_date = post.date_local.date()

            if count < 3 and post_date < date_threshold:                           # to bypass pinned post
                count += 1
                continue
            if post.date_local.date() < date_threshold:
                break
            if post_date >= date_threshold and post_date not in processed_dates and not post.is_video:
                processed_dates.add(post_date)

                # Download the image from the post
                loader.download_post(post, target=username)

                # Get the list of downloaded image files
                image_files = [f for f in os.listdir(username) if f.endswith(".jpg")]
                if image_files:
                    selected_image = image_files[0]

                    # Process the selected image and save it
                    image_path = os.path.join(username, selected_image)
                    enhanced_image_path = os.path.join(folder_path, selected_image)
                    image = Image.open(image_path)
                    image.save(enhanced_image_path)

                    print(f"Processed image from {post_date}")
                    break  # Stop processing after selecting one image per date

        print("Scraping completed.")
    except instaloader.exceptions.ProfileNotExistsException:
        print("Profile does not exist.")

for username in usernames:
    print(username)
    scrape_images(username)
```

```python
from PIL import Image
import requests
import os


# Specify the path to the folder containing images
image_folder_path = "/content/drive/MyDrive/output_imageSet"

# Get a list of image files in the folder
image_files = [f for f in os.listdir(image_folder_path) if f.lower().endswith(".jpg")]
# Loop through the image files and process each image
for image_file in image_files:
    image_path = os.path.join(image_folder_path, image_file)
    # Open the image using PIL
    image = Image.open(image_path)

    wear = {}
    top_dict = {}
    bottom_dict = {}
    foot_dict = {}

    # print("*"*50)
    scores = classifier(image,
                    candidate_labels = ['men','women'])
    sum = 0
    #topwear
    level2label=json_data_topwear[scores[0]['label']]
    scores2 = classifier(image,
                    candidate_labels = level2label)
    for i in range(5):
      top_dict[scores2[i]['score']] = scores2[i]['label']
      sum += scores2[i]['score']
    wear["top_dict"] = sum
    # print("*"*10)

    sum = 0
    #bottomwear
    level2label=json_data_bottomwear[scores[0]['label']]
    scores2 = classifier(image,
                    candidate_labels = level2label)
    for i in range(5):
      bottom_dict[scores2[i]['score']] = scores2[i]['label']
      sum += scores2[i]['score']
    wear["bottom_dict"] = sum
    # print("*"*10)

    sum = 0
    #footwear
    level2label=json_data_footwear[scores[0]['label']]
    scores2 = classifier(image,
                    candidate_labels = level2label)
    for i in range(5):
      foot_dict[scores2[i]['score']] = scores2[i]['label']
      sum += scores2[i]['score']
    wear["foot_dict"] = sum
    # print("*"*10)


    # print(wear)
    # print(top_dict)
    # print(bottom_dict)
    # print(foot_dict)

    # RANKING THE OUTFITS
    sorted_wear = dict(sorted(wear.items(), key=lambda item: item[1]))
    the_two_ = list(sorted_wear.keys())[1:]

    if (the_two_[0]=="top_dict" and the_two_[1]=="bottom_dict") or (the_two_[0]=="bottom_dict" and the_two_[1]=="top_dict"):
      combined_two = {**top_dict, **bottom_dict}
    elif (the_two_[0]=="foot_dict" and the_two_[1]=="bottom_dict") or (the_two_[0]=="bottom_dict" and the_two_[1]=="foot_dict"):
      combined_two = {**foot_dict, **bottom_dict}
    else:
      combined_two = {**foot_dict, **top_dict}

    # print(combined_two)
    ranked = [combined_two[key] for key in sorted(combined_two.keys(), reverse=True)[:2]]

    for r in ranked:
      print(r)
print("Image processing completed")
```
